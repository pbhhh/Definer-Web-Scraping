```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from ankr import AnkrWeb3
```

```python
ankr_w3 = AnkrWeb3()

block_list = []

for i in range(41799268,41799270):
    blocks = ankr_w3.query.get_blocks(
    blockchain="polygon",
    from_block=i,
    to_block=i,
    desc_order=True,
    include_logs=True,
    include_txs=True,
    decode_logs=True)
    
#start here    if
    
    df_1 = pd.DataFrame(blocks[0])
    df_1

    df_2 = pd.DataFrame(df_1[1][19])
    df_2

    logs_1 = pd.DataFrame(df_2[17].to_list())
    logs_1

    flat_list = [item for sublist in logs_1[1].tolist() for item in sublist]
    flat_list

    logs_2 = pd.DataFrame(flat_list)
    logs_2

    col1 = [tple[1] for tple in logs_2[[0,4,9]][0]]

    col2 = [tple[1] for tple in logs_2[[0,4,9]][4]]

    col3 = [tple[1] for tple in logs_2[[0,4,9]][9]]
    col3

    col4 = [tple.name for tple in col3]

    block_list.append({'pool address': col1,
                       'transaction hash': col2,
                       'method': col4})

block_list1 = pd.DataFrame(block_list)
block_list1

pool_address = [item for sublist in block_list1['pool address'] for item in sublist]
pool_address
txs_hash = [item for sublist in block_list1['transaction hash'] for item in sublist]
txs_hash
method = [item for sublist in block_list1['method'] for item in sublist]
method

log = pd.DataFrame({'pool address': pool_address,
                    'transaction hash': txs_hash,
                    'method': method})
log

swap = log[log.method.str.contains('Swap')]
swap
```
